```python
import pandas as pd

file_PATH = './data/conseil_docente.csv'
df = pd.read_csv(file_PATH)
df.drop(columns=['G.libero', 'Ore tot', 'Strutture'], inplace=True)

len(df.columns)

new_rows = []
for index, row in df.iterrows():
    if not pd.isna(row['Compresenze']):
        index = '+'.join(sorted( (row['Identificativo'], row['Compresenze']) ))
        new_rows.append(
            [
                index, row['Materia'], row['Classi'], row['N.Ore'], None
            ]
        )
print(new_rows)
new_df = pd.DataFrame(new_rows, columns=df.columns)
df = pd.concat([df, new_df], ignore_index=True)

df.drop(df[df['Compresenze'].notna()].index, inplace=True)
df.sort_values(by = 'Identificativo', inplace=True)
df.to_csv('./data/conseil_docente_cleaned.csv', index=False)
```
