```python
from collections import defaultdict
import pandas as pd

file_path = './data/docente_classes.csv'

df = pd.read_csv(file_path)
professors = set(df['Identificativo'].unique())
classes = set(df['Classi'].unique())
total_resources = dict()
for name, group in df.groupby('Identificativo'):
    resources_by_profesor = []
    for index, row in group.iterrows():
        teacher = row['Identificativo']
        materia = row['Materia']
        Classi = row['Classi']
        nb_hours = row['N.Ore']
        if not pd.isna(nb_hours):
            nb_hours = int(nb_hours)
        else:
            nb_hours = 0
        resources_by_profesor.extend(int(nb_hours)*[(materia, Classi)])
    total_resources[name] = resources_by_profesor
```
