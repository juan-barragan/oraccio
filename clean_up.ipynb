```python
import pandas as pd

orario_path = './data/spagnolo.csv'
orario = pd.read_csv(orario_path)

orario.sort_values(by = 'teacher', inplace=True)
orario.set_index('teacher', inplace=True)

orario.columns = [
    f'{d}{h}' for d in ['LUN', 'MAR', 'MER', 'GIO', 'VEN'] for h in [8, 9, 10, 11, 12, 13, 14]
]
```

```python
from dataframe import logic
from collections import defaultdict

new_rows = dict()
for col in orario.columns:
    d = logic.duplicated_indices_in_column(orario, col)
    if len(d)>0:
        d = sorted(d)
        key = f'{d[0]} {d[1]}'
        classe = orario.loc[d[0], col]
        orario.loc[d[0], col] = ''
        orario.loc[d[1], col] = ''
        if key in new_rows:
            new_rows[key][col] = classe
        else:
            new_row = pd.Series('', index=orario.columns)
            new_row[col] = classe
            new_row.name = key
            new_rows[key] = new_row
if new_rows:
        orario = pd.concat([orario, pd.DataFrame(new_rows.values())], axis=0)
```

```python
orario.sort_index(inplace=True)

orario.to_csv('spagnolo_imp.csv')
```
